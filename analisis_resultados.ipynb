```python
import pandas as pd

# Lista de variables de experimentación
experimentaciones = [
    experimentacion1, experimentacion2, experimentacion3, experimentacion4, experimentacion5,
    experimentacion6, experimentacion7, experimentacion8, experimentacion9, experimentacion10, original
]

# Inicializar lista para guardar los resultados
resultados = []

# Iterar sobre cada DataFrame y calcular los resultados
for i, df in enumerate(experimentaciones, start=1):
    # Calcular el promedio de la función objetivo
    promedio_funcion_objetivo = df["Funcion Objetivo"].mean()
    
    # Tomar valores de las columnas deseadas (primer fila como referencia)
    cant_labo = df["Cant Labo"].iloc[0]
    cant_radio = df["Cant Radio"].iloc[0]
    nginec1 = df["NGinec1"].iloc[0]
    nginec2 = df["NGinec2"].iloc[0]
    nginec3 = df["NGinec3"].iloc[0]
    
    # Agregar los resultados a la lista
    resultados.append({
        "Experimentación": i,
        "Promedio Función Objetivo": promedio_funcion_objetivo,
        "Cant Labo": cant_labo,
        "Cant Radio": cant_radio,
        "NGinec1": nginec1,
        "NGinec2": nginec2,
        "NGinec3": nginec3
    })

# Crear un DataFrame con los resultados
tabla_resultados = pd.DataFrame(resultados)

# Mostrar la tabla
print(tabla_resultados)

# Opcional: guardar en un archivo Excel o CSV
# tabla_resultados.to_excel("Comparacion_Experimentaciones.xlsx", index=False)
# tabla_resultados.to_csv("Comparacion_Experimentaciones.csv", index=False)
```

```python
import scipy.stats as stats

# Filtrar las experimentaciones seleccionadas
exp_11 = original["Funcion Objetivo"]
exp_1 = experimentacion1["Funcion Objetivo"]
exp_2 = experimentacion2["Funcion Objetivo"]
exp_3 = experimentacion3["Funcion Objetivo"]

# Realizar el test ANOVA
f_stat, p_value = stats.f_oneway(exp_11, exp_1, exp_2, exp_3)

# Imprimir los resultados
print(f"Estadístico F: {f_stat}")
print(f"Valor p: {p_value}")

# Tomar una decisión
if p_value < 0.05:
    print("Hay diferencias significativas entre las medias de las experimentaciones.")
else:
    print("No hay diferencias significativas entre las medias de las experimentaciones.")
```

```python
import pandas as pd
import scipy.stats as stats

# Crear el DataFrame de resultados
resultados = pd.DataFrame({
    'Experimentación': [1, 2, 3, 4, 5, 6, 7],
    'Promedio_Funcion_Objetivo': [46.284056, 44.750128, 37.431556, 43.667556, 42.864056, 41.207133, 40.624203]
})

# Filtrar los datos para la experimentación 1 y las demás
experimento_1 = resultados[resultados['Experimentación'] == 1]['Promedio_Funcion_Objetivo']
otras_experimentaciones = resultados[resultados['Experimentación'] != 1]['Promedio_Funcion_Objetivo']

# Aplicar un ANOVA de una vía (un solo factor) para comparar las medias
f_statistic, p_value = stats.f_oneway(experimento_1, otras_experimentaciones)

# Mostrar los resultados
print("Estadística F:", f_statistic)
print("Valor p:", p_value)

# Evaluación de la hipótesis
alpha = 0.05  # Nivel de significancia
if p_value < alpha:
    print("Rechazamos la hipótesis nula: la media de la Experimentación 1 es diferente a las demás.")
else:
    print("No rechazamos la hipótesis nula: no hay evidencia suficiente para afirmar que la media de la Experimentación 1 sea diferente a las demás.")
```
